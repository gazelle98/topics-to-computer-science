### Import libraries

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

import seaborn as sns
import pandas as pd
```

### Import data from file and do some data cleaning

```python
# Reading data from file and see the head of it
data_raw = pd.read_csv('train_data.csv')
data = data_raw.copy(deep=True)
data.head()
```

```python
# Check the data overally
data.describe()
```

```python
data.info()
```

It seems that there are some rows that their Experience column are not positive integer so I remove those rows in order to increase the correctness of our data.


```python
data = data[data['Experience'] >= 0]
data.describe()
```

```python
sns.countplot(x='Personal Loan', data=data)
```

The ZIP code column must not be in integer format because the object of ZIP code is not related to numbers and It will bias our model incorrectly. 

```python
data['ZIP Code'] = data['ZIP Code'].astype(str)
data.info()
```

```python
sns.countplot(x='Education', data=data)
```

```python
sns.catplot(x='Personal Loan', col='Education', kind='count', data=data)
```

```python
feature_cols = ['Age', 'Experience', 'Education', 'Income', 'CCAvg', 'Family', 'Mortgage', 'Personal Loan']
temp_data = data[feature_cols]
sns.pairplot(data=temp_data, hue='Personal Loan')
```

```python
# There is no NULL value in the dataset.
data.isnull().sum()
```

### Feature Engineering and creating new features

```python
# It must be a good feature to calculate the ratio of money spent to the income of the customer in a year
data['Spend/Income'] = (data['CCAvg'] * 12) / data['Income']

# Another feature is to consider if the customer has family or not
data['Is Alone'] = 1
data['Is Alone'].loc[data['Family'] > 1] = 0

# Another feature calculate the whole money that the customer spent with bank services
data['All Spent'] = data['Experience'] * data['CCAvg']

feature_cols += ['Is Alone', 'Spend/Income', 'All Spent']
```

```python
data.head()
```

### Training a model with K-Nearest Neighbor Algorithm

```python
y = data['Personal Loan']
X = data[feature_cols]

# Split the data to two part in order to have train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Use GridSearchCV to get the best score from the K-Nearest Algorithm
neighbors = list(range(1,30))
param_grid = {
    'n_neighbors': neighbors,
    'weights': ['uniform', 'distance'], 
    'metric': ['euclidean', 'manhattan']
}
gs = GridSearchCV(KNeighborsClassifier(), param_grid, cv=10, scoring='accuracy')
gs_result = gs.fit(X_train, y_train)
gs.best_score_
```

```python
# Running the best estimated model
neigh = KNeighborsClassifier(**gs.best_params_)
neigh.fit(X_train, y_train)
predicted = neigh.predict(X_test)
accuracy_score(y_test, predicted)
```

As we see above, the rate of success for the future campaign increases a little bit. So we try another algorithm.

### Training a model with Random Forest Classifier Algorithm 

```python
forest_model = RandomForestClassifier(random_state=0)
param_grid = {
    'n_estimators': [10, 1000],
    'max_features': ['auto', 'sqrt', 'log2'], 
    'max_depth': [2, 3, 4, 5, 6, 7],
    'criterion': ['gini', 'entropy']
}
cv = GridSearchCV(estimator=forest_model, param_grid=param_grid, cv=5)
cv.fit(X_train, y_train)
cv.best_score_
```

```python
rfc = RandomForestClassifier(**cv.best_params_)
rfc.fit(X_train, y_train)
predicted2 = rfc.predict(X_test)
print('Test accuracy score: ', accuracy_score(y_test, predicted2))
```

The second algorithm gets a better score. So this is our model.
